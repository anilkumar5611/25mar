```python
# Flight Price Dataset:

# q1    Load the flight price dataset and examine its dimensions.
# How many rows and columns does the dataset have?
```

```python
import pandas as pd

# Load the dataset
flight_data = pd.read_csv('flight_price.csv')

# Dimensions of the dataset
rows, cols = flight_data.shape
print("Number of rows:", rows)
print("Number of columns:", cols)
```

```python
# q2    What is the distribution of flight prices in the dataset? 
# Create a histogram to visualize the distribution.
```

```python
import matplotlib.pyplot as plt

# Histogram for flight prices
plt.hist(flight_data['Price'], bins=20, color='skyblue', edgecolor='Black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')
plt.show()
```

```python
# q3=====What is the range of prices in the dataset? 
# What is the minimum and maximum price?
```

```python
min_price = flight_data['Price'].min()
max_price = flight_data['Price'].max()
price_range = max_price - min_price
print("Minimum price:", min_price)
print("Maximum price:", max_price)
print("Price range:", price_range)
```

```python
# q4----How does the price of flights vary by airline? 
# Create a boxplot to compare the prices of different airlines.
```

```python
import seaborn as sns

# Boxplot for flight prices by airline
plt.figure(figsize=(12, 6))
sns.boxplot(x='Airline', y='Price', data=flight_data)
plt.xticks(rotation=45)
plt.title('Flight Prices by Airline')
plt.show()
```

```python
# q5-----Are there any outliers in the dataset? Identify any potential outliers using a boxplot 
# and describe how they may impact your analysis.
```

```python
# Boxplot to identify outliers
plt.figure(figsize=(8, 6))
sns.boxplot(flight_data['Price'], orient='v')
plt.ylabel('Price')
plt.title('Identifying Outliers in Flight Prices')
plt.show()
```

## Google Playstore Dataset:

```python
# Load the dataset
playstore_data = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Dimensions of the dataset
rows_playstore, cols_playstore = playstore_data.shape
print("Number of rows:", rows_playstore)
print("Number of columns:", cols_playstore)
```

```python
# Boxplot for app ratings by category
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Rating', data=playstore_data)
plt.xticks(rotation=90)
plt.title('App Ratings by Category')
plt.show()
```

```python
# Check for missing values
missing_values = playstore_data.isnull().sum()
print(missing_values)
```

```python
# Scatter plot for app size and rating
plt.figure(figsize=(8, 6))
plt.scatter(playstore_data['Size'], playstore_data['Rating'], alpha=0.5)
plt.xlabel('Size')
plt.ylabel('Rating')
plt.title('Relationship between App Size and Rating')
plt.show()
```

```python

# Frequency table for top 10 most popular apps
top_10_apps = playstore_data[['App', 'Installs']].sort_values(by='Installs', ascending=False).head(10)
print(top_10_apps)
```
